```python
# Function to convert pounds to kilograms]
def lbs_to_kg(lbs):
    return round(lbs * 0.453592, 2)

# Read number of students
N = int(input("Enter the number of students: "))

# Initialize an empty list to store weights in lbs
weights_lbs = []

# Read weights in lbs from user input
print(f"Enter the weights of {N} students in lbs:")
for _ in range(N):
    weight = float(input())
    weights_lbs.append(weight)

# Initialize an empty list to store weights in kg
weights_kg = []

# Convert each weight to kg and add to the weights_kg list
for weight in weights_lbs:
    weights_kg.append(lbs_to_kg(weight))

# Print the original weights in lbs and the converted weights in kg
print(f"Original weights in lbs: {weights_lbs}")
print(f"Converted weights in kg: {weights_kg}")
```
